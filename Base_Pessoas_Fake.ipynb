```python
import pandas as pd
import random
from faker import Faker

# Inicialize o gerador de dados fictícios com localização no Brasil
fake = Faker('pt_BR')

# Listas vazias para armazenar os dados
nomes = []
sexos = []
datas_nascimento = []
cidades_nascimento = []
estados_nascimento = []
idades = []
telefones = []

# Gere 10 linhas de dados fictícios
for _ in range(1000):
    nome = fake.first_name() + " " + fake.last_name()
    sexo = random.choice(['Masculino', 'Feminino'])
    data_nascimento = fake.date_of_birth(tzinfo=None, minimum_age=18, maximum_age=80)
    cidade_nascimento = fake.city()
    estado_nascimento = fake.state_abbr()
    idade = fake.random_int(min=18, max=80, step=1)
    telefone = fake.phone_number()

    nomes.append(nome)
    sexos.append(sexo)
    datas_nascimento.append(data_nascimento)
    cidades_nascimento.append(cidade_nascimento)
    estados_nascimento.append(estado_nascimento)
    idades.append(idade)
    telefones.append(telefone)

# Crie o DataFrame
data = {
    'Nome': nomes,
    'Sexo': sexos,
    'Data de Nascimento': datas_nascimento,
    'Cidade de Nascimento': cidades_nascimento,
    'Estado de Nascimento': estados_nascimento,
    'Idade': idades,
    'Telefone': telefones
}

df = pd.DataFrame(data)

# Exiba o DataFrame
df.head()
```

```python
df.to_csv('Base_pessoas_fake.csv',index=False)
```
